```python
def yeeers(lizt):
    for i in range(len(lizt)): #the names of the docs
        doc = lizt[i]+'.csv'
        data = pd.read_csv(doc, low_memory = False)
        KK = list(data['_id'].str[:4])
        kk = [int(i) for i in KK] 
        print(kk[:10])
        data.insert(1, 'Year', kk)
        #print(data.head())
      #  print(data.head())
        data.to_csv('hmk3'+ doc, index=False)
    return 
```

```python
liz2 = ['hmk3jan.csv',
      'hmk3fsumm.csv',
       'hmk3depperf.csv',
       'hmk3candiv.csv',
     'hmk3arrperf.csv']
```

```python
li = []
for i in liz2:
    a = pd.read_csv(i)
    a.drop(['Unnamed: 0'], axis = 1, inplace = True)
    li.append(a)
    a.to_csv(i, index = False)
li
```

```python
for i in range(1,5):
    li[i].drop(['Year'], axis = 1, inplace = True)
```

```python
len(li)
```

```python
df_final = reduce(lambda left,right: pd.merge(left,right,on='_id'), li)
df_final.head()
```
